#### Grouping the null values together

```python
trainDf.columns[trainDf.isnull().any()] 

nullDf = trainDf.isnull().sum()/len(trainDf)*100
nullDf = nullDf[nullDf>0]
nullDf.sort_values(inplace=True, ascending=False)
nullDf
```

#### Replacing the null values with none on places where it woudnt affect the prices

```python
DfWithoutNull = ["Alley", "MasVnrType", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "PoolQC", "Fence", "MiscFeature"]
for i in DfWithoutNull:
    trainDf[i].fillna("none", inplace=True)
```

#### Checking again if null values are there

```python
nullDf2 = trainDf.isnull().sum()/len(trainDf)*100
nullDf2 = nullDf2[nullDf2>0]
nullDf2.sort_values(inplace=True, ascending=False)
nullDf2
```

#### Removing Outliers

```python
def removeOutliers(x,y):
    q1 = x[y].quantile(0.25)
    q3 = x[y].quantile(0.75)
    value = q3-q1
    lower_value  = q1-1.5*value
    higer_value = q3+1.5*value
    out= x[(x[y]<higer_value) & (x[y]>lower_value)]
    return out
```
